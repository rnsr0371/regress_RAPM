```python
import pandas as pd
import scipy.stats
import statsmodels.api as sm
import numpy as np
```

```python
stats_202021=pd.read_csv("traditional_stats_202021.csv")
stats_202021
```

```python
stats_201920=pd.read_csv("traditional_stats_201920.csv")
stats_201920
```

```python
stats_201819=pd.read_csv("traditional_stats_201819.csv")
stats_201819
```

```python
stats_201718=pd.read_csv("traditional_stats_201718.csv")
stats_201718
```

```python
stats_201617=pd.read_csv("traditional_stats_201617.csv")
stats_201617
```

```python
stats=pd.concat([stats_202021,stats_201920,stats_201819,stats_201718,stats_201617])
stats
```

```python
stats=stats.groupby("Player").mean()
stats
```

```python
RAPM=pd.read_csv("5year_RAPM.csv")
RAPM=RAPM.loc[:,["Player","Rapm"]]
RAPM
```

```python
data=RAPM.merge(stats,on="Player")
data
```

```python
Y=data["Rapm"]
X=sm.add_constant(data.loc[:,["PTS","REB","AST","STL","BLK","TOV"]])
model=sm.OLS(Y,X)
results=model.fit()
```

```python
results.summary()
```

```python
data["PTS_Z"]=scipy.stats.zscore(data["PTS"])
data["REB_Z"]=scipy.stats.zscore(data["REB"])
data["AST_Z"]=scipy.stats.zscore(data["AST"])
data["STL_Z"]=scipy.stats.zscore(data["STL"])
data["BLK_Z"]=scipy.stats.zscore(data["BLK"])
data["TOV_Z"]=scipy.stats.zscore(data["TOV"])
```

```python
data.describe()
```

```python
Y=data["Rapm"]
X=sm.add_constant(data.loc[:,["PTS_Z","REB_Z","AST_Z","STL_Z","BLK_Z","TOV_Z"]])
model=sm.OLS(Y,X)
results=model.fit()
```

```python
results.summary()
```

```python
data["PTS_48"]=data.loc[:,"PTS"]/data.loc[:,"MIN"]*48
data["REB_48"]=data.loc[:,"REB"]/data.loc[:,"MIN"]*48
data["AST_48"]=data.loc[:,"AST"]/data.loc[:,"MIN"]*48
data["STL_48"]=data.loc[:,"STL"]/data.loc[:,"MIN"]*48
data["BLK_48"]=data.loc[:,"BLK"]/data.loc[:,"MIN"]*48
data["TOV_48"]=data.loc[:,"TOV"]/data.loc[:,"MIN"]*48

data["FGA_48"]=data.loc[:,"FGA"]/data.loc[:,"MIN"]*48
data["FTA_48"]=data.loc[:,"FTA"]/data.loc[:,"MIN"]*48
```

```python
Y=data["Rapm"]
X=sm.add_constant(data.loc[:,["PTS_48","REB_48","AST_48","STL_48","BLK_48","TOV_48"]])
model=sm.OLS(Y,X)
results=model.fit()
results.summary()
```

```python
data["PTS_48_Z"]=scipy.stats.zscore(data["PTS_48"])
data["REB_48_Z"]=scipy.stats.zscore(data["REB_48"])
data["AST_48_Z"]=scipy.stats.zscore(data["AST_48"])
data["STL_48_Z"]=scipy.stats.zscore(data["STL_48"])
data["BLK_48_Z"]=scipy.stats.zscore(data["BLK_48"])
data["TOV_48_Z"]=scipy.stats.zscore(data["TOV_48"])

data["MIN_Z"]=scipy.stats.zscore(data["MIN"])
```

```python
Y=data["Rapm"]
X=sm.add_constant(data.loc[:,["PTS_48_Z","REB_48_Z","AST_48_Z","STL_48_Z","BLK_48_Z","TOV_48_Z"]])
model=sm.OLS(Y,X)
results=model.fit()
results.summary()
```
